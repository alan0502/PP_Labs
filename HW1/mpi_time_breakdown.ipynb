# MPI Time Breakdown
這個 Notebook 將輸入總執行時間，並根據分析結果把時間分成：
- I/O
- Communication
- CPU (剩餘時間)

```python
import pandas as pd

# 從 CSV 讀取
df = pd.read_csv("/home/lab/PP_Labs/HW1/profile/ori/N1_n1/rank_0_mpi_event_sum.csv")


# 檢查前幾列
#print(df.head())
df
```

```python
# 定義分類
io_ops = ['MPI_File_read_at', 'MPI_File_write_at', 'MPI_File_open', 'MPI_File_close']
comm_ops = ['MPI_Sendrecv', 'MPI_Barrier']

# 計算 I/O 與 Communication 時間 (轉換成秒)
io_time = df[df['Name'].isin(io_ops)]['Total Time (ns)'].sum() / 1e9
comm_time = df[df['Name'].isin(comm_ops)]['Total Time (ns)'].sum() / 1e9

# 輸入總時間 (單位改成秒)
total_time = 25.53

# CPU time = 總時間 - I/O - Communication
cpu_time = total_time - io_time - comm_time

io_time, comm_time, cpu_time
```

```python
# 轉換成秒，並輸出更易讀的結果
result = {
    'I/O (s)': io_time / 1e9,
    'Communication (s)': comm_time / 1e9,
    'CPU (s)': cpu_time / 1e9,
    'Total (s)': total_time / 1e9
}

result
```
